```python
import numpy as np
import matplotlib.pyplot as plt
# Sample data
np.random.seed(0)
steps = np.random.choice([-1, 1], 100, replace=True)
original_values = np.cumsum(steps)
# Compute a moving average
window_size = 5
averages = np.convolve(original_values, np.ones(window_size)/window_size, mode='valid')
# Plotting

top_range = original_values + 0.5
bot_range = original_values - 0.5


plt.figure(figsize=(10, 6))
plt.plot(top_range, color='lightgray', alpha=0.5)
plt.plot(original_values, label='Moving Average', color='blue', linestyle='-')
plt.plot(bot_range, color='lightgray', alpha=0.5)
plt.fill_between(range(window_size - 1, 100), original_values[window_size-1:], top_range[window_size-1:], color='blue', alpha=0.2)
plt.fill_between(range(window_size - 1, 100), original_values[window_size-1:], bot_range[window_size-1:], color='blue', alpha=0.2)

plt.title('Original Values vs. Moving Average')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# Import required libraries
import matplotlib.pyplot as plt

# Data for the bar graph
bars = ['FedLE-OneShot', 'FedLE-MultiShot', 'FEDAVG-Bonly', 'FEDAVG-withB']
heights = [90, 90, 78, 50]

# Set the positions of the bars on the x-axis
x_pos = [i for i, _ in enumerate(bars)]

# Create a bar graph
plt.bar(x_pos, heights)

# Assign labels to the x-axis
plt.xticks(x_pos, bars)

# Set the title and labels for the axes
plt.xlabel("Battery")
plt.ylabel("Methods")
plt.title("CIFAR-Clients battery level at round 50")

# Display the bar graph
plt.show()
```

```python
color = ["red","green","yellow"]


for ids, one_filename in enumerate(glob.glob("MNIST CNN on Non-IID Dataset-Bonly*.csv")):
  rounds = []
  train_loss = []
  with open(one_filename, 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
      rounds.append(int(row.get('round')))
      train_loss.append(float(row.get('test_accuracy')))
    plt.plot(rounds, train_loss, label=str(color[ids]))
    plt.xlabel('Number of Rounds')
    plt.ylabel('Test Accuracy')

print(rounds)

print(train_loss)
```
