```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
fig = plt.figure(figsize=(10, 6))
data_withb = pd.read_csv("NIID2-WithB-AvR_test_accuracy-0.8.csv")
data_bonly = pd.read_csv("NIID2-BONLY-AvR_test_accuracy-0.8.csv")
data_cucb = pd.read_csv("NIID2-CUCB-AvR_test_accuracy-0.8.csv")
data_fedle = pd.read_csv("NIID2-FEDLE-AvR_test_accuracy-0.8.csv")
data_poc = pd.read_csv("NIID2-POC-AvR_test_accuracy-0.8.csv")

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data_withb['round'], data_withb['average_test_accuracy'], color='blue', label='WithB')
plt.plot(data_bonly['round'], data_bonly['average_test_accuracy'], color='green', label='BOnly')
plt.plot(data_cucb['round'], data_cucb['average_test_accuracy'], color='red', label='CUCB')
plt.plot(data_fedle['round'], data_fedle['average_test_accuracy'], color='purple', label='FEDLE')
plt.plot(data_poc['round'], data_poc['average_test_accuracy'], color='orange', label='POC')

plt.xlabel('Round')
plt.ylabel('Average Test Accuracy')
plt.title('Average Test Accuracy vs Round')
plt.legend(loc='lower right')
plt.grid(True)

fig.savefig("NIID2-0.8-average_test_accuracy.pdf", dpi=1200, format='pdf')

plt.show()
```
