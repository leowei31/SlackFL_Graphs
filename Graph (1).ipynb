```python
# Import required libraries
import matplotlib.pyplot as plt

# Data for the bar graph
bars = ['FedLE-OneShot', 'FedLE-MultiShot', 'FEDAVG-Bonly', 'FEDAVG-withB']
heights = [90, 90, 78, 50]

# Set the positions of the bars on the x-axis
x_pos = [i for i, _ in enumerate(bars)]

# Create a bar graph
plt.bar(x_pos, heights)

# Assign labels to the x-axis
plt.xticks(x_pos, bars)

# Set the title and labels for the axes
plt.xlabel("Battery")
plt.ylabel("Methods")
plt.title("CIFAR-Clients battery level at round 50")

# Display the bar graph
plt.show()
```

```python
color = ["red","green","yellow"]


for ids, one_filename in enumerate(glob.glob("MNIST CNN on Non-IID Dataset-Bonly*.csv")):
  rounds = []
  train_loss = []
  with open(one_filename, 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
      rounds.append(int(row.get('round')))
      train_loss.append(float(row.get('test_accuracy')))
    plt.plot(rounds, train_loss, label=str(color[ids]))
    plt.xlabel('Number of Rounds')
    plt.ylabel('Test Accuracy')

print(rounds)

print(train_loss)
```
